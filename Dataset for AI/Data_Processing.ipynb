### renaming columns

We also found that the column names are also drifted in df1 onwards, while the data enteries in the respective cells are always the same. So we rename columns of all the dfs based on the first df: df0

```python
# prompt: change df1,df2,.., df6 column names to df0's

import pandas as pd

# Assuming df0, df1, ..., df6 are already defined pandas DataFrames

# Get the column names from df0
new_column_names = df0.columns.tolist()

# Iterate through the dataframes and rename their columns
for i in range(1, 7):
  try:
    globals()[f'df{i}'].columns = new_column_names
  except KeyError:
    print(f"DataFrame df{i} not found.")
  except AttributeError:
    print(f"df{i} is not a DataFrame")
```

```python
# prompt: concat df0,to df6 without considering coulumn names in df1 to df6

import pandas as pd

# Assuming df0, df1, ..., df6 are already defined DataFrames

# Concatenate the DataFrames, ignoring column names in df1 to df6
concatenated_df = pd.concat([df0] + [df.reset_index(drop=True) for df in [df1, df2, df3, df4, df5, df6]], ignore_index=True)


# Now concatenated_df holds the combined data
concatenated_df

```

## Dealing with non-numerical columns using Encoding

I have found that some columns have non-numrical values. It is important to identify these coulmns to treat them accordingly for further analysis.


```python
nonnumerical_columns = df.select_dtypes(exclude=[np.number]).columns.tolist()
print("Non-numerical columns in df:", nonnumerical_columns)
```

Since in non-numerical columns, there are features and target columns, we will separate them using the exert knowledge that columns from cost onwards are targets, which needed to be convert.

```python

features_to_convert = nonnumerical_columns[:-16] # Exclude the last 16 columns which are likely target columns
targets_to_convert = nonnumerical_columns[-16:]  # Last 16 columns are likely target columns
print("Features to convert:", features_to_convert)
print("Targets to convert:", targets_to_convert)
```

```python
# Identify categorical and continuous non-numerical columns, then plot only categorical ones

categorical_cols = []
continuous_cols = []

for col in nonnumerical_columns:
    unique_count = df[col].nunique()
    if unique_count < 20:
        categorical_cols.append(col)
    else:
        continuous_cols.append(col)

print("Categorical columns:", categorical_cols)
print("Continuous or high-cardinality columns:", continuous_cols)

```
